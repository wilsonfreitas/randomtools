# The design of random variables

In my opinion, the comprehension of how random numbers are generated is quite essential. Nowadays, we have at hand a great amount of libraries offering all sorts of random number generators already implemented. It's cheap and useful, mainly when you just want to do a simple test. It doesn't worth to come accross a deep understanding of how a random number generator is implemented.
I would say that it's always relevant to know exactly what are you doing, how and why, but in some cases I also understand that you have to belive that the implementation is suitable for your needs and all you have to do is just going on.
The Python language comes with the `random` module which brings some random number generators for various distributions (most popular: beta, normal, power law, exponential and many others) and also the powerful Mersenne Twister generator which generate random uniformly floats in the semi-open range [0.0, 1.0).
Besides that random module, the Python scientific community brought the Scipy framework which has the package stats that covers a wide range of distributions and consequently its random number generators.

**randomtools** is a Python module mainly concerned with the educational purpose of let people know how the undercovered algorithms work.
It isn't its intention being fast or even replace any other random number generator package.
Instead, here we can see how much can be done with simple functions and a single random number generator $U(0,1)$.



```python
%pylab inline
```

```python
import randomtools as rt
import numpy as np
from hyptests import jarque_bera_test
from statsmodels.stats.stattools import jarque_bera
```

```python
rng = np.random.random
unif = rt.uniform(rng, -1.0, 1.0)
unif
```

```python
ht = plot([unif() for r in range(100)])
```

```python
unif = rt.uniform(rng, 0.0)
unif
```

```python
hlines(1, 0, 1, colors='r')
ht = hist([unif(1.0) for r in range(1000)], bins=50, normed=True)
```

```python
randn = rt.clt(rng, N=20)
```

```python
vlines(0, 0, 0.6, colors='r')
ht = hist([randn() for r in range(1000)], bins=50, normed=True)
```
