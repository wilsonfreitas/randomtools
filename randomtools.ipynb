## Uniform random variates

Let's start with the simplest distribution, the **Continuous Uniform Distribution**.

$$
U(a,b)
$$

where $a$ and $b$ define that that variate is inside the interval $[a,b)$.
So, if I have a variable which follows that distribution I don't expect outcomes outside that range.
Here it follows a implementation of the `uniform` generator.

```python
def uniform(rand, a, b):
    return a + (b-a) * rand()
```

As you can see it is fairly simple and that is the code implemented in `randomtools.uniform`.
It expands the `rand` spectrum and then shifts it.

> **Note on `randomtools` functions:**
> 
> Despite that simple implementation above, the `randomtools.uniform`, like many others `randomtools` functions, returns a function that returns a generator, the random number generator itself.
> Most `randomtools` functions work as a factory which build a preset random number generator.
> It uses features like decorators, closures and generators, fairly diffused among Python practitioners.
> The use of that features allowed me to focus on the algorithmic implementation without interference of collateral effects of delivery mechanism.

To use `uniform` we pass `rng` and the limits.

````python
runif = rt.uniform(rng, -1.0, 1.0)
runif
````

As we can see it returns a function and that function is a preset random number generator.
`runif` have been created to generate random numbers following $U(-1,1)$ and to use it you must pass it to the `list` constructor providing the quantity you want to generate.

````python
list(runif(5))
````

As you can see we've got 5 random numbers.
In the plot below we can observe the result equivalent to what we have got before.
The only difference is on the specified range for the generated numbers, accordingly to what have been set.

> **Note on the need of `list` constructor**
>
> The reason why `runif` returns a generator instead of sequence is because I decided to avoid the creation of that sequence until I realy need it.
> In simulations, like discrete event simulations or Monte Carlo simulations, I don't need the sequence at once I use the random numbers bit by bit, and in the end, I evaluate my result.
> This is a concept called Lazy Evaluation.

In the chart bellow we observe that our square have been expanded.

````python
scatter(list(runif(1000)), list(runif(1000)), color='green')
scatter(rng(1000), rng(1000), color='red');
````

### Compuging $\pi$

````python
from itertools import izip, starmap

pi = lambda m, n: 4*m/float(n)
inside = lambda x, y: x*x + y*y < 1

def compute_pi(N):
    m = reduce(lambda x,y: x+y, starmap(inside, izip(runif(N), runif(N))))
    return pi(m, N)

print(['{0:<6}: {1:.4f}'.format(x, compute_pi(x)) for x in (500, 5000, 50000)])
````

## Gaussian random variates

````python
rnorm = rt.clt(rng, 20)
rnorm
````

````python
ylim(-4, 4)
xlim(-4, 4)
scatter(list(rnorm(1000)), list(rnorm(1000)), color='red');
````

````python
randn = rt.boxmuller(rng)
ylim(-3,3)
xlim(-3,3)
ht = plot([r for r in randn(1000)], [r for r in randn(1000)], '.r')
````

````python
randn = rt.marsaglia(rng)
ylim(-3,3)
xlim(-3,3)
ht = plot([r for r in randn(1000)], [r for r in randn(1000)], '.r')
````

````python
randn = rt.moro(rng)
ylim(-3,3)
xlim(-3,3)
ht = plot([r for r in randn(1000)], [r for r in randn(1000)], '.r')
````

````python
randn = rt.moro(rng)
smp = [r for r in randn(1000)]

kde = sm.nonparametric.KDE(smp)
kde.fit()

ht = hist(smp, normed=True, bins=25, color='red')
plot(kde.support, kde.density, lw=2, color='black');
````
