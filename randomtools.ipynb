# randomtools

Sometimes it is important to know the steps taken in order to generate random numbers. I would say that it's always important, but in some cases I also understand that you have to belive that the random number generator algorithm is suitable for your needs and all you have to do is go on.
**randomtools** is a Python module mainly concerned with the educational purpose of let people know how the undercovered algorithms work.
It isn't its intention being fast or even replace any other random number generator package.
Instead, here we can see how much can be done with simple functions and a single random number generator $U(0,1)$.



```python
%pylab inline
```

```python
import randomtools as rt
import numpy as np
from hyptests import jarque_bera_test
from statsmodels.stats.stattools import jarque_bera
```

```python
rng = np.random.random
unif = rt.uniform(rng, -1.0, 1.0)
unif
```

```python
ht = plot([unif() for r in range(100)])
```

```python
unif = rt.uniform(rng, 0.0)
unif
```

```python
hlines(1, 0, 1, colors='r')
ht = hist([unif(1.0) for r in range(1000)], bins=50, normed=True)
```

```python
randn = rt.clt(rng, N=20)
```

```python
vlines(0, 0, 0.6, colors='r')
ht = hist([randn() for r in range(1000)], bins=50, normed=True)
```
